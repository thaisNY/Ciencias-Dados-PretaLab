### ExercГ­cio 1 вЂ“ Cadastro de alunas (usando `faker` e `random`)

- Imagine que vocГЄ Г© responsГЎvel por organizar uma lista de alunas para um curso.

- Use a biblioteca `faker` para gerar pelo menos 30 nomes de alunas.

- Para cada aluna, crie tambГ©m um nГєmero de matrГ­cula aleatГіrio (por exemplo, entre 1000 e 9999).

- Salve os dados em um arquivo chamado alunas.txt, com o formato: `Nome - MatrГ­cula`





```python
!pip install faker
```

```python

from faker import Faker
import random

# InicializaГ§ГЈo
fake = Faker('pt_BR')
nome_alunos = []
matriculas = []

# 1. Gerando os dados (30 alunas)
for i in range(30):
    nome = fake.name()
    nome_alunos.append(nome)

    matricula = random.randint(1000, 9999)
    matriculas.append(matricula)

# 2. Imprimindo as listas completas para conferir
print("Lista de Nomes:", nome_alunos)
print("Lista de MatrГ­culas:", matriculas)

# 3. Salvando no arquivo alunas.txt
with open("alunas.txt", "w", encoding="utf-8") as arquivo:
    for i in range(30):
        # Escreve no formato: Nome - MatrГ­cula
        linha = f"{nome_alunos[i]} - {matriculas[i]}\n"
        arquivo.write(linha)

print("\nArquivo 'alunas.txt' criado com sucesso!")
```

### ExercГ­cio 2 вЂ“ Analisando nГєmeros de uma pesquisa (usando random)

VocГЄ estГЎ ajudando em uma pesquisa que deseja coletar 200 respostas numГ©ricas (valores entre 1 e 1000).

1. Gere esses nГєmeros aleatГіrios e salve-os em um arquivo numeros.txt.

2. Escreva um programa que leia o arquivo e descubra: Qual foi a menor resposta registrada? Qual foi a maior resposta registrada?

Esse tipo de anГЎlise Г© importante para entender extremos em pesquisas ou levantamentos de dados.

```python
numeros = []
for i in range(200):
    numero = random.randint(1, 1000)
    numeros.append(numero)

maior = max(numeros)
menor = min(numeros)

print(f"Maior nГєmero: {maior}")
print(f"Menor nГєmero: {menor}")

with open("numeros.txt", "w", encoding="utf-8") as arquivo:
  for i in range(200):
    linha = f"{numeros[i]}\n"
    arquivo.write(linha)
```

### ExercГ­cio 4 вЂ“ Criando um mГіdulo de utilidades para arquivos

Agora que jГЎ temos vГЎrios arquivos (`alunas.txt`, `numeros.txt`, `frases.txt`), vamos criar um **mГіdulo** arquivos.py para facilitar a manipulaГ§ГЈo de arquivos.

1. Esse mГіdulo deve conter as seguintes funГ§Гµes:

- `contar_caracteres(nome_arquivo)` в†’ devolve o nГєmero total de caracteres do arquivo.

- `contar_palavras(nome_arquivo)` в†’ devolve a quantidade total de palavras no arquivo.

- `ler_conteudo(nome_arquivo, qtd_linhas)` в†’ lГЄ e devolve apenas as primeiras N linhas do arquivo (N Г© definido pelo usuГЎrio).

2. Teste cada funГ§ГЈo do mГіdulo com os arquivos criados (`alunas.txt`, `numeros.txt`, `frases.txt`).

рџ‘‰ Isso mostra como modularizar cГіdigo ajuda a organizar programas maiores.

```python
def contar_caracteres(nome_arquivo):
  with open("frases.txt", "r", encoding="utf-8") as f:
    conteudo = f.read() # LГЄ o arquivo INTEIRO como uma string
    total = len(conteudo)
  print(f"O arquivo tem {total} caracteres (incluindo espaГ§os e \n).")


print(contar_caracteres("frases.txt"))
```

```python
def contar_palavras(nome_arquivo):
  with open("frases.txt", "r", encoding="utf-8") as f:
    conteudo = f.read().split()
    total = len(conteudo)
  print(f"O arquivo tem {total} palavras.")


print(contar_palavras("frases.txt"))
```
