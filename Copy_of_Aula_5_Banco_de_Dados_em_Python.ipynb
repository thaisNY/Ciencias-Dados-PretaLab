# Prática CRUD com SQLite3 em Python

```python
import sqlite3
```

## 1. Criando conexão com o banco SQLite

```python
con = sqlite3.connect("ecormmece.db")
cur = con.cursor()
```

## 2. Criando tabelas

```python
cur.execute("""
CREATE TABLE IF NOT EXISTS usuarios (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    nome TEXT NOT NULL,
    email TEXT UNIQUE NOT NULL,
    senha TEXT NOT NULL,
    data_criacao TIMESTAMP DEFAULT CURRENT_TIMESTAMP
);
""")

cur.execute("""
CREATE TABLE IF NOT EXISTS produtos (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    nome TEXT NOT NULL,
    preco REAL NOT NULL,
    estoque INTEGER NOT NULL
);
""")

cur.execute("""
CREATE TABLE IF NOT EXISTS pedidos (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    usuario_id INTEGER NOT NULL,
    data_pedido TIMESTAMP DEFAULT CURRENT_TIMESTAMP,
    status TEXT DEFAULT 'Pendente',
    FOREIGN KEY(usuario_id) REFERENCES usuarios(id)
);
""")

cur.execute("""
CREATE TABLE IF NOT EXISTS itens_pedido (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    pedido_id INTEGER NOT NULL,
    produto_id INTEGER NOT NULL,
    quantidade INTEGER NOT NULL,
    preco_unitario REAL NOT NULL,
    FOREIGN KEY(pedido_id) REFERENCES pedidos(id),
    FOREIGN KEY(produto_id) REFERENCES produtos(id)
);
""")

con.commit()
```

## 3. Inserindo dados (CREATE)

```python
#Adicionando usuários
usuarios = [
    ("João Silva", "joao@email.com", "123456"),
    ("Maria Souza", "maria@email.com", "abcdef"),
    ("Carlos Pereira", "carlos@email.com", "senha123")
]

cur.executemany("INSERT INTO usuarios (nome, email, senha) VALUES (?, ?, ?)", usuarios)

#Adicionando produtos
produtos = [
    ("Azeite Andorinha", 39.90, 20),
    ("Papel Higienico", 20, 10),
    ("Suco de Laranje", 15, 3)
]

cur.executemany("INSERT INTO produtos (nome, preco, estoque) VALUES (?, ?, ?)", produtos)

con.commit()
```

## 4. Consultando dados (READ)

```python
pedidos = [
  (1, 'Pendente'),
(2, 'Confirmado'),
(3, 'Enviado'),
(1,'Entregue'),
(2, 'Pendente'),
(3, 'Confirmado'),
(1, 'Enviado'),
(2, 'Pendente'),
(3, 'Confirmado'),
(1, 'Entregue'),

]

cur.executemany("INSERT INTO pedidos (usuario_id, status) VALUES (?, ?)", pedidos)

con.commit()
```

```python
#Consultando usuários
print("Usuários:")
for row in cur.execute("SELECT * FROM usuarios"):
    print(row)

#Consultando produtos
print("\nProdutos:")
for row in cur.execute("SELECT * FROM produtos"):
    print(row)


#Consultando pedidos
print("\nPedidos:")
for row in cur.execute("SELECT * FROM pedidos"):
    print(row)
```

## 5. Atualizando dados (UPDATE)

```python
cur.execute("UPDATE produtos SET estoque = estoque - 1 WHERE id = 1")
con.commit()

print("\nEstoque atualizado:")
for row in cur.execute("SELECT * FROM produtos"):
    print(row)
```

## 6. Deletando dados (DELETE)

```python
cur.execute("DELETE FROM usuarios WHERE id = 3")
con.commit()

print("\nUsuários após exclusão:")
for row in cur.execute("SELECT * FROM usuarios"):
    print(row)
```

## 7. Consultas com agregação

```python
print("\nPreço médio dos produtos:")
for row in cur.execute("SELECT AVG(preco) FROM produtos"):
    print(row)

print("\nTotal de produtos cadastrados:")
for row in cur.execute("SELECT COUNT(*) FROM produtos"):
    print(row)
```

```python
#Thais
import sqlite3

# Conectando ao banco (ou criando um em memória para teste)
con = sqlite3.connect(":memory:")
cur = con.cursor()

# 1. Resetando e Criando o Schema
# Usamos DROP para você poder rodar o código várias vezes sem erro de UNIQUE
cur.execute("DROP TABLE IF EXISTS categorias_produtos")

cur.execute("""
CREATE TABLE categorias_produtos (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    nome TEXT UNIQUE NOT NULL,
    descricao TEXT,
    status BOOLEAN NOT NULL
);
""")

# 2. Definindo os dados (DML)
categorias_dados = [
    ("Lego", "brinquedo de montar", True),
    ("Monopoly", "jogo de tabuleiro financeiro", True),
    ("UNO", "jogo de cartas", True)
]

# 3. Executando a Query de Inserção
# O executemany percorre a lista e insere cada item nos placeholders (?)
cur.executemany("""
    INSERT INTO categorias_produtos (nome, descricao, status)
    VALUES (?, ?, ?)
""", categorias_dados)

# Salva as alterações no banco
con.commit()

# 4. Listando e verificando os resultados (DQL)
print("Conteúdo da tabela categorias_produtos:")
print("-" * 30)

for row in cur.execute("SELECT * FROM categorias_produtos"):
    print(f"ID: {row[0]} | Nome: {row[1]} | Status: {row[3]}")

# Fechando a conexão
con.close()
```
