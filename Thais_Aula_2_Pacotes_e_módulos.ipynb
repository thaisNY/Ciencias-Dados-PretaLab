# Módulos e Pacotes em Python

## Por que módulos e pacotes?
Reaproveitar código acelera o desenvolvimento e evita "reinventar a roda".

Um **módulo** é um arquivo `.py` com definições (funções, classes, variáveis).

Um **pacote** é uma pasta com módulos (geralmente contém um `__init__.py`).

**Importar** significa "trazer" nomes de outro módulo para usar aqui.

## Biblioteca padrão
Python vem com muitos módulos úteis já como padrão, alguns exemplos são:
- `math`: funções matemáticas
- `random`: aleatoriedade
- `os` / `pathlib`: arquivos e diretórios
- `datetime`: datas e horários
- `json`: JSON é o formato padrão para APIs e arquivos de configuração. Em Python, usamos o módulo nativo json para converter strings em dicionários e vice-versa. Referência: https://www.json.org/json-en.html

## Sintaxe de import
Para usar um módulo utilizamos a palavra `import`:

- `import modulo`
- `import modulo as apelido`
- `from modulo import nome1, nome2`
- `from modulo import *` (pode causar conflitos e não é uma boa prática)

### Funções úteis:
- `max(lista)` → retorna o maior valor.

- `min(lista)` → retorna o menor valor.

- `sum(lista)` → soma os valores (ótimo se quiser calcular média).

- `len(lista)` → quantidade de elementos.

```python
import random
lista_100 = []

for i in range (5):
  n = random.sample(range(1, 100), 1)
  lista_100.append(n)

lista_ordenada = sorted(lista_100)
print(max(lista_ordenada))
print(min(lista_ordenada))
print(lista_ordenada)
```

## Pacotes `requests` e `BeautifulSoup`

Para extração de dados.

```python
import requests
from bs4 import BeautifulSoup

def extrair_noticia(url):
  # Fazendo a requisição
  resposta = requests.get(url)

  if resposta.status_code == 200:
      # Parse do HTML
      soup = BeautifulSoup(resposta.text, "html.parser")

      # Pegando o título (geralmente está em <h1>)
      titulo = soup.find("h1")

      print("Título da notícia:")
      print(titulo.get_text() if titulo else "Não encontrado")

      # Pegando os primeiros parágrafos (geralmente em <p>)
      paragrafos = soup.find_all("p")
      print("\nPrimeiros parágrafos:")
      for p in paragrafos[:3]:
          print("-", p.get_text())
  else:
      print("Erro ao acessar a página:", resposta.status_code)
```

```python
url_bbc = "https://www.bbc.com/portuguese/articles/cp98231le9mo"
extrair_noticia(url_bbc)
```

```python
url_g1 = "https://g1.globo.com/ciencia/noticia/2025/09/16/mumias-fumaca-cientistas-apontam-as-mumificacoes-mais-antigas-do-mundo-asia.ghtml"
extrair_noticia(url_g1)
```

Exemplo:

Plaintext
Arroz: 10
Feijão: 5
Macarrão: 8
Azeite: 2
Sal: 4
Parte 2: O Programa
Escreva um código que execute as seguintes etapas:

Leitura Inicial: Leia o arquivo e exiba o estoque atual de forma organizada para o usuário.

Entrada do Usuário: Pergunte ao usuário qual item ele deseja adicionar e qual a quantidade.

Atualização: O programa deve verificar se o item já existe:

Se existir: Some a nova quantidade à quantidade antiga.

Se não existir: Adicione uma nova linha com o produto e a quantidade.

Escrita: Salve a lista atualizada de volta no arquivo estoque.txt.

```python
with open("desafio-pessoal.txt", "w", encoding="utf-8") as d:
    d.write("Arroz:10\n")
    d.write("Feijão:5\n")
    d.write("Macarrão:8\n")
    d.write("Azeite:2\n")
    d.write("Sal:4\n")



with open("desafio-pessoal.txt", "r", encoding="utf-8") as d:
  conteudo = d.read()
  print(conteudo)


while True:
  resp = input("Quer adicionar algo na lista? (s/n): ").lower()
  if resp != "n":
      novo_item = input("Digite o nome do novo produto: ")
      quantidade = int(input("Digite a quantidade: "))
      if novo_item not in conteudo:
        with open("desafio-pessoal.txt", "a", encoding="utf-8") as arquivo:
            arquivo.write(f"{novo_item}:{quantidade}\n")
            print(f"'{novo_item}' foi adicionado com sucesso!")
      else:
        print("Item já existe! Vamos atualizar...")

        # PASSO 1: Ler o que já existe e guardar numa "tabela" (dicionário)
        estoque_temporario = {}
        with open("desafio-pessoal.txt", "r", encoding="utf-8") as arquivo:
            for linha in arquivo:
                nome_no_arquivo, qtd_no_arquivo = linha.strip().split(":")
                estoque_temporario[nome_no_arquivo] = int(qtd_no_arquivo)

        # PASSO 2: Somar a nova quantidade no item que já existia
        estoque_temporario[novo_item] += quantidade

        # PASSO 3: Salvar tudo de volta, limpando o arquivo antigo
        with open("desafio-pessoal.txt", "w", encoding="utf-8") as arquivo:
            for nome, qtd in estoque_temporario.items():
                arquivo.write(f"{nome}:{qtd}\n")

        print(f"Sucesso! Agora você tem {estoque_temporario[novo_item]} de {novo_item}.")


  else:
      print("Obrigado por usar o programa!")
      with open("desafio-pessoal.txt", "r", encoding="utf-8") as d:
        conteudo = d.read()
        print(conteudo)
        break
```
